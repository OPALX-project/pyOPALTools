# 1D example

```python
# true model
def model(x):
    return 0.5 * x[:, 0]**3 + x[:, 0]**2 - 2.5 + x[:, 0]
```

```python
from uq import UQ
import numpy as np
import matplotlib.pyplot as plt
from bootstrap import plot_confidence


x_train = 2 + np.random.random((200, 1))
x_test = 2 + np.random.random((40, 1))
y_train = model(x_train)

n_boot = 100      # number of bootstrapping iterations
sample = 'random' # boostrapping: random (with replacement), loo (leave-one-out), lpo (leave-p-out)
alpha  = 0.95     # CI interval
order  = 2
uq = UQ(order)

y_train, y_pred, y_lo, y_up = uq.confidence_interval(x_train, y_train, alpha=alpha,
                                                     n_boot=n_boot, sample=sample)

plt.figure(figsize=(16, 9))
plt = plot_confidence(y_train, y_pred, y_lo, y_up, alpha=alpha)

y_test = uq.predict(x_test)
y_true = model(x_test)
plt.scatter(y_true, y_test, marker='.', color='red', label='test points')

plt.legend()

plt.show()
plt.close()
```

# 2D example

```python
# true model
def model(x):
    return x[:, 0] * x[:, 1] - 2.0 * x[:, 1]**3 + x[:, 0] ** 2 * x[:, 1] - 3.0 + x[:, 0]
```

```python
from uq import UQ
import numpy as np
import matplotlib.pyplot as plt
from bootstrap import plot_confidence


x_train = np.hstack([2 + np.random.random((200, 1)),
                     2 + np.random.random((200, 1))])


x_test = np.hstack([2 + np.random.random((50, 1)),
                    2 + np.random.random((50, 1))])

y_train = model(x_train)

n_boot = 100      # number of bootstrapping iterations
sample = 'random' # boostrapping: random (with replacement), loo (leave-one-out), lpo (leave-p-out)
alpha  = 0.95     # CI interval
order  = 1
uq = UQ(order)

y_train, y_pred, y_lo, y_up = uq.confidence_interval(x_train,
                                                     y_train, alpha=alpha, n_boot=n_boot,
                                                     sample='random')

plt.figure(figsize=(16, 9))
plt = plot_confidence(y_train, y_pred, y_lo, y_up, alpha=alpha)

y_test = uq.predict(x_test)
y_true = model(x_test)
plt.scatter(y_true, y_test, marker='.', color='red', label='test points')

plt.legend()

plt.show()
plt.close()
```
