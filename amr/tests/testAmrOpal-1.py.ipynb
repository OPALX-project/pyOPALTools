## Load Dataset

```python
from amr import AmrOpal

ao = AmrOpal()


ao.load_file(dirname='/home/matthias/Documents/projects/pyOPALTools/amr/tests/plt0000000380/',
             showme=False)
```

## Projection Plot

```python
ao.projection_plot(axis='z', field='rho', unit='C/m**3', zoom=1.0, color='gray')
```

## Slice Plot

```python
ao.slice_plot(normal='z', field='Ex', unit='V/m',
              zoom=1.0, color='gray')
```

```python
ao.slice_plot(normal='z', field='electric_field_y', unit='V/m',
              zoom=1.0, color='gray')
```
