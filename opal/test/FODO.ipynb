# Example: FODO

## Load the Gantry.stat file

```python
from opal.opal import load_dataset

ds = load_dataset('./FODO/', fname='FODO.stat')
```

## Set the plotting style

```python
from opal.visualization.styles import load_style

load_style('jupyter')
```

## Envelope Plot

```python
plt = ds.plot_envelope(lfile='./FODO/FODO.lattice')

plt.show()
```
